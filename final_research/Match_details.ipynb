```python
ODI_matches = pd.concat([ODI_2019,ODI_2020,ODI_2021])
```

```python
ODI_matches_filter =ODI_matches.drop(['Margin','Ground','Scorecard'],axis = 1)
```

```python
ODI_matches_filter = ODI_matches_filter.reset_index(drop=True)
ODI_matches_filter.loc[ODI_matches_filter.Winner == ODI_matches_filter['Team 1'],'Result']=1
ODI_matches_filter.loc[ODI_matches_filter.Winner == ODI_matches_filter['Team 2'], 'Result']=0
ODI_matches_filter.drop(['Winner'],axis=1, inplace=True)

ODI_matches_filter.head()
```

```python
ODI_matches_filter['Result'] = ODI_matches_filter['Result'].fillna(0).astype(int)
```

```python
ODI_matches_filter['Match Date'] = pd.to_datetime(ODI_matches_filter['Match Date']).dt.year
```
